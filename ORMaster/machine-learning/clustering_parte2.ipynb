## Clusterização Hierárquica

Temos dois tipos de clusterização hierárquica:

* __Aglomerativo__: Essa é uma abordagem "bottom-up": cada observação inicia em em seu próprio cluster, e pares de clusters são aglomerados de forma hierárquica.
* __Divisível__: Essa é uma abordagem "top-down": todas as observações iniciam em um único cluster e divisões são feitas de maneira recursiva de acordo com a hierarquia.

O caso de uso mais comum da clusterização hierárquica é o método "bottom-up". Focaremos, então, no entendimento dessa variante. 

Para a implemetação do algoritmo, criamos as funções ```get_clusters()```, ```dist()```, ```get_closest()```, ```agglomerate()``` e ```plot_result()```.

A primeira função recebe os pontos e os centróides dos clusters e associa cada ponto a um cluster baseado na distância Euclidiana. A função ```dist()``` calcula a distância Euclidiana entre dois pontos e é utilizada pela função ```get_closest()``` para identificar os dois clusters mais próximos. A função ```agglomerate()``` retorna a nova posição dos clusters. A última função (```plot_result()```) cria um gráfico para facilitar a visualização dos resultados do algoritmo.

````python
import matplotlib.pyplot as plt
import numpy as np

# associa cada conto ao centróide mais próximo 
def get_clusters(x, y, c_x, c_y):
    d = []
    for i in range(len(c_x)):
        x_ = np.full((1, len(x)), c_x[i])[0]
        y_ = np.full((1, len(y)), c_y[i])[0]
        d.append(((x - x_) ** 2 + (y - y_) ** 2) ** 1/2)
    d = np.array(d)
    return [np.argmin(d[:, j]) for j in range(len(x))]

# Calcula a distância Euclidiana entre dois pontos
def dist(x1, x2, y1, y2):
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5 

# Identifica os dois clusters mais próximos 
def get_closest(x, y, done_list=[]):
    dist_matrix = np.array([[ dist(x[i], x[j], y[i], y[j]) 
                             if i != j and i not in done_list and j not in done_list else 1e7 for j in range(len(x))] for i in range(len(x))])
    i = np.argmin(dist_matrix) // len(x)
    j = int(round((np.argmin(dist_matrix) / len(x) - np.argmin(dist_matrix) // len(x)) * len(x), 0))
    return (i, j)

# Faz o processo de aglomeração
def agglomerate(x, y, c_x, c_y):
    new_c_x = []
    new_c_y = []
    i, j = get_closest(c_x, c_y)
    c_x = np.append(c_x, np.array([np.mean([c_x[i], c_x[j]])]))
    c_y = np.append(c_y, np.array([np.mean([c_y[i], c_y[j]])]))
    if i != j:
        c_x = np.delete(c_x, [i, j])
        c_y = np.delete(c_y, [i, j])
    else:
        c_x = np.delete(c_x, [i])
        c_y = np.delete(c_y, [i])
    return (c_x, c_y)

# mostra a etapa do algoritmo na forma de gráficos
def plot_result(x, y, c_x, c_y, clusters=None):
    plt.figure(figsize=(15, 5))
    plt.subplot(121)
    plt.scatter(x, y)
    plt.scatter(c_x, c_y, color='red', marker='v')
    plt.legend(['pontos', 'cluster'])
    if clusters is not None:
        plt.subplot(122)
        for c in clusters:
            for i, (value_x, value_y) in enumerate(zip(x, y)):
                if clusters[i] == c:
                    plt.plot([value_x, c_x[c]], [value_y, c_y[c]], '--', color='black')
        plt.scatter(x, y)
        plt.scatter(c_x, c_y, color='red', marker='v')
        plt.text(1, 7, f'k = {len(c_x)}')
````

Consideraremos o exemplo abaixo para a demonstração do algoritmo.

````python
x = np.array([1.5, 2.2, 1.1, 3.2, 4.5, 5.6, 6.4, 9.1, 10.5, 9.1, 3.7, 9.6, 5.7, 2.8, 5.8, 8.4, 1.9, 2, 2.4, 9.2])
y = np.array([2.2, 1.3, 3.4, 1.2, 4.1, 3.7, 4.5, 5.3, 6.0, 7.2, 2.3, 6.7, 4.7, 2.6, 4.1, 5.9, 2.7, 2, 1.9, 5.9])
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
````

Primeiro, associamos cada ponto a um cluster próprio.

````python
c_x = np.array(list(x))
c_y = np.array(list(y))
clusters = get_clusters(x, y, c_x, c_y)
plot_result(x, y, c_x, c_y, clusters)
````

Executando o código abaixo repeditas vezes obtemos os passos do algorítimo, reduzindo em 1 o número de clusters em cada iteração.

````python
c_x, c_y = agglomerate(x, y, c_x, c_y)
clusters = get_clusters(x, y, c_x, c_y)
plot_result(x, y, c_x, c_y, clusters)
````

Para obter a clusterização para um $k$ específico, utilizamos o código abaixo:

````python
k = 3

c_x = np.array(list(x))
c_y = np.array(list(y))
while len(c_x) > k:
    c_x, c_y = agglomerate(x, y, c_x, c_y)
clusters = get_clusters(x, y, c_x, c_y)
plot_result(x, y, c_x, c_y, clusters)
````

Utilizaremos os dados contidos no arquivo ```exemplo1.csv``` para testar o algoritmo em um caso real utilizando o pacote ```sklearn```.

````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('dados/exemplo1.csv')
plt.figure(figsize=(10, 7))
plt.xlabel('idade')
plt.ylabel('conta corrente')
plt.scatter(df.idade, df.conta_corrente)
````

Utilizaremos a função abaixo para gerar a visualização dos clusters.

````python
def plot(df, column_name):
    plt.figure(figsize=(10, 7))
    plt.xlabel('idade')
    plt.ylabel('conta corrente')
    for c in sorted(df[column_name].unique()):
        plt.scatter(df[df[column_name] == c].idade, df[df[column_name] == c].conta_corrente)
    plt.legend([f'cluster {c}' for c in sorted(df[column_name].unique())])
````

Faremos a normalização dos dados utilizando o método ```MinMaxScaler``` e importaremos o algoritmo de clusterização hierárquico ```AgglomerativeClustering```. 

Para $k=3$, temos:

````python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering


X = MinMaxScaler().fit_transform(df[['idade', 'conta_corrente']])
df['h_cluster3'] = AgglomerativeClustering(n_clusters=3).fit_predict(X)
plot(df, 'h_cluster3')
````

Experimentando com o número de clusters igual a 4, temos um resultado semelhante ao obtido com o KMeans.

````python
df['h_cluster4'] = AgglomerativeClustering(n_clusters=4).fit_predict(X)
plot(df, 'h_cluster4')
````

A função abaixo permite a plotagem do resultado de múltiplas clusterizações para conparação.

````python
def plot_clusters(df, column_list):
    plt.figure(figsize=(15, 6))
    for i, column_name in enumerate(column_list):
        plt.subplot(1, len(column_list), i + 1)
        plt.xlabel('idade')
        plt.ylabel('conta corrente')
        for c in sorted(df[column_name].unique()):
            plt.scatter(df[df[column_name] == c].idade, df[df[column_name] == c].conta_corrente)
        plt.legend([f'cluster {c}' for c in sorted(df[column_name].unique())])
        plt.title(column_name)
````

Faremos a comparação visual dos métodos Kmeans e Hierárquico para $k=3$ e $k=4$

````python
from sklearn.cluster import KMeans


df['km_cluster3'] = KMeans(n_clusters=3, random_state=5).fit_predict(X)
df['km_cluster4'] = KMeans(n_clusters=4, random_state=3).fit_predict(X)
plot_clusters(df, ['h_cluster3', 'km_cluster3'])
````

Para $k=3$ percebemos uma pequena diferença na fronteira entre o cluster 0 e o cluster 2.

````python
plot_clusters(df, ['h_cluster4', 'km_cluster4'])
````

Para $k=4$, percebemos pequenas diferenças nas fronteiras entre os clusters 0 e 2 e os clusters 2 e 3.
